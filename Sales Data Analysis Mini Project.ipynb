## Task 4

#### Find out the monthly revenue and analyze the findings
1. Create a column with Year and month field only give a suitable name to it
2. Create a separate dataset from the data which will have two columns , one which is created in step a of this task and second the monthly sales/revenue
3. Plot this dataset
4. Describe your analysis for the monthly revenue observations.

```python
# Create a column with Year and month field only give a suitable name to it

df["Year-Month"] = df['Order Date'].dt.strftime('%Y-%m')
#df.head()


# Create a separate dataset from the data which will have two columns , one which is created in step a of this task and second the monthly sales/revenue

results_months = pd.DataFrame(df.groupby('Year-Month').sum()['Sales'])
months = [month for month, df in df.groupby('Year-Month')]
# results_months

# Plot this dataset

plt.figure(figsize=(15,5))
plt.plot(months,results_months['Sales'], color = '#b80045')
plt.xticks(months, rotation='vertical', size = 8)
plt.title('Monthly Revenue')
plt.ylabel('Sales')
plt.xlabel('Month')
plt.grid()
plt.show()
```

###### Describe your analysis for the monthly revenue observations.


The lowest sales was observed in Feb. 2014, while the highest sales was observed in Nov. 2017. As observed in the graph we can say that the sales is not consistent it keeps on fluctuating.

## Task 6

#### Finding out the monthly growth rate and analyse the findings
1. Create a column in the monthly revenue dataset for monthly growth rate
2. Find out the maximum monthly growth rate
3. Plot the findings
4. Describe the findings

```python
# Create a column in the monthly revenue dataset for monthly growth rate

# Monthly_Revenue = pd.DataFrame(results_months['Sales'])
results_months['Last_Month'] = np.roll(results_months['Sales'],1)
results_months = results_months.drop(results_months.index[0])
results_months['Growth Rate'] = (results_months['Sales']/results_months['Last_Month'])-1
# results_months
#Now, since the monthly growth rate of first month cannot be found since we donot have it's previous month value.
#By default, python has put the last value in the column and rolled up over there. SO we need to remove that.
```

```python
# Plot the findings
results = results_months.drop(columns = ["Sales", "Last_Month"])
results['Months'] = results.index
results.reset_index(drop=True, inplace=True)
plt.figure(figsize=(15,5))
plt.bar(results['Months'],results['Growth Rate']*100, color = '#b80045')
plt.xticks(results['Months'], rotation='vertical', size = 8)
plt.ylabel('% Growth')
plt.xlabel('Year-Months')
plt.title("Monthly Growth Rate")
#plt.grid()
plt.show()
```

```python
# Find out the maximum monthly growth rate
results_months.sort_values(by=['Growth Rate'], inplace=True, ascending=False)
results_months.head(1)
```

###### Describe the findings

From the graph, we can tell that, we get highest growth rate in March 2014. In some of the year we can see that we get negative growth rate as well. Also, in some months, we have even got null growth rate. If you see in the graph that we have not considered January 2014 because, before that month we were not having any earlier data to compare, so we have neglected that month.

## Task 10

#### Finding out the majority and minority customer cities on basis of
1. Number of customers
2. Sales value
3. Number of quantity sold

```python
# Based on Number of customers

Nos_Customer = df.groupby(["City"]).size().reset_index(name='Customer_No') 
Nos_Customer.sort_values(by=['Customer_No'], inplace=True, ascending=True)
#print(Nos_Customer)

A = np.array(Nos_Customer)
#print(np.shape(A)) 

# For minority
print('The Minority cities based on Number of Customer [City Name, Customer] are as follows: \n')

i=0
count=0
while i<len(A) :
    if A[i,1]== A[:,1].min() :
        count = count + 1
    i=i+1


# print(count)
# print(A[69,1])
# print(A[70,1])

j=0
while j<count :
    P=A[j,:]
    j=j+1
    print(P)


# For majority
print('\n\n')
print('The Majority cities based on Number of Customer [City Name, Customer] are as follows: \n')
k=len(A)-1
temp=0
while k>0 :
    if A[k,1]== A[:,1].max() :
        temp = temp + 1
    k=k-1
    
# print(temp)


while temp>0 :
    print(A[len(A)-temp,:])
    temp =temp-1
```

```python
# Based on Sales value

City_Sales = df[['City','Sales']].copy()
City_Sales = pd.DataFrame(City_Sales.groupby('City').sum() ['Sales'])
# City_Sales
City_Sales.sort_values(by=['Sales'], inplace=True, ascending=True)
# print(City_Sales)

B = City_Sales.reset_index().values
# print(B)
# print(np.shape(B))

# For minority
print('The Minority cities based on Sales Value [City Name, Sales] are as follows: \n')
a=0
c=0
while a<len(B) :
    if B[a,1]== B[:,1].min() :
        c = c + 1
    a=a+1

x=0
while x<c :
    Q=B[x,:]
    x=x+1
    print(Q)
    

# For majority
print('\n\n')
print('The Majority cities based on Sales Value [City Name, Sales] are as follows: \n')
l=len(B)-1
t=0
while l>0 :
    if B[l,1]== B[:,1].max() :
        t = t + 1
    l=l-1

while t>0 :
    print(B[len(B)-t,:])
    t =t-1
```

```python
# Based on Number of quantity sold
City_Quantity = df[['City','Quantity']].copy()
City_Quantity = pd.DataFrame(City_Quantity.groupby('City').sum() ['Quantity'])
# City_Quantity
City_Quantity.sort_values(by=['Quantity'], inplace=True, ascending=True)
# print(City_Quantity)

C = City_Quantity.reset_index().values
# print(C)
# print(np.shape(C))

# For minority
print('The Minority cities based on Number of quantity sold [City Name, Quantity] are as follows: \n')
m=0
v=0
while m<len(C) :
    if C[m,1]== C[:,1].min() :
        v = v + 1
    m=m+1

y=0
while y<v :
    S=C[y,:]
    y=y+1
    print(S)
    

# For majority
print('\n\n')
print('The Majority cities based on Number of quantity sold [City Name, Quantity] are as follows: \n')
n=len(C)-1
z=0
while n>0 :
    if C[n,1]== C[:,1].max() :
        z = z + 1
    n=n-1

while z>0 :
    print(C[len(C)-z,:])
    z =z-1
```

#### Find out the most and least sold product category from the store

```python
# Value based

Category_Value = df.groupby(["Category"]).size().reset_index(name='Total Value Sold') 
Category_Value.sort_values(by=['Total Value Sold'], inplace=True, ascending=True)
# Category_Value
D = np.array(Category_Value)
# print(np.shape(D))
# print(D)

# For least sold
print('The least sold product category from the store [Category, Total Value Sold] are as follows: \n')
r=0
e=0
while r<len(D) :
    if D[r,1]== D[:,1].min() :
        e = e + 1
    r=r+1

u=0
while u<e :
    R=D[u,:]
    u=u+1
    print(R)
    
    
# For most sold
print('\n\n')
print('The most sold product category from the store [Category, Total Value Sold] are as follows: \n')
w=len(D)-1
g=0
while w>0 :
    if D[w,1]== D[:,1].max() :
        g = g + 1
    w=w-1

while g>0 :
    print(D[len(D)-g,:])
    g =g-1
```

```python
# Quantity based

Category_Quantity = df[['Category','Quantity']].copy()
Category_Quantity = pd.DataFrame(Category_Quantity.groupby('Category').sum() ['Quantity'])
# Category_Quantity
Category_Quantity.sort_values(by=['Quantity'], inplace=True, ascending=True)
# print(Category_Quantity)

E = Category_Quantity.reset_index().values
# print(np.shape(E))
# print(E)

# For least sold
print('The least sold product category from the store [Category, Quantity] are as follows: \n')
rr=0
ee=0
while rr<len(E) :
    if E[rr,1]== E[:,1].min() :
        ee = ee + 1
    rr=rr+1

uu=0
while uu<ee :
    T=E[uu,:]
    uu=uu+1
    print(T)


# For most sold
print('\n\n')
print('The most sold product category from the store [Category, Quantity] are as follows: \n')
ww=len(E)-1
gg=0
while ww>0 :
    if E[ww,1]== E[:,1].max() :
        gg = gg + 1
    ww=ww-1

while gg>0 :
    print(E[len(E)-gg,:])
    gg =gg-1
```
